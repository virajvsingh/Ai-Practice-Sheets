```python
### Importing the library pandas, loading csv file and prints the columns mentioned

import pandas as pd

df = pd.read_csv('pokemon_data.csv')
#can also use a text file using read_csv with a delimiter defined 

# df =  pd.read_excel('data_file.xlsx')

print(df.columns)   #prints the names of columns
print(df['Name'])   #prints the values in the column name
print(df[['Name','HP','Type 1']])   #prints the values of columns of [Name, HP, Type 1]
```

```python
### Extracting values from the data files

print(df.head(4)[['Name','Type 1','Type 2','HP','Attack','Defense']]) # prints 4 rows from the top including given columns
print(df.tail(4)[['Name','Type 1','Type 2','HP','Attack','Defense']]) # prints 4 rows from the bottom including given columns

#data_frame.iloc[row1:row2,col1:col2,step]
print(df.iloc[1:4,1:6])
print(df.iloc[8,1])
```

```python
### Iterating through the data 

for index,row in df.iterrows():
    print(index,row[['Name','Type 1']])
    # can go through the values one by one and execute the commands
    
### Working with big data or making smaller data frames

for df_chunk in pd.read_csv('pokemon_data.csv', chunksize = 5):
    print(df_chunk)
    break
    
```

```python
### Describing the data 
df.describe()

```

```python
### Sorting values

df.sort_values(['Name','HP'], ascending = False) # sorts on the basis of name and then HP
```

```python
### Playing with columns

new_df = pd.read_csv('pokemon_data.csv') #read data

# Add a new column called Efficiency
new_df['Efficiency'] = new_df['HP']/new_df['Attack']
print(new_df.sort_values('Efficiency',ascending =  False))
#df.dump(columns = ['Efficiency'])   adds the column to the main file


# Add a new column adding all the values from column 4 to 9(not included)
new_df['Total'] = new_df.iloc[:,4:9].sum(axis = 1)
print(new_df.sort_values('Total',ascending =  False))

# axis in pandas vs cartesian system
# X axis =  1 (row)
# Y axis = 0  (column)
# Z axis = 2  (depth)
```

```python
cols = list(df.columns.values) # creates a list of column names
print(cols)
```

```python
### Filtering data

filtered_data = df.loc[(df['Type 1'] == 'Bug') & (df['Type 2'] == 'Poison')]
filtered_data.reset_index(drop = True) #Drops the original indexes

print(filtered_data)

# locates all the data that given string
print(df.loc[df['Name'].str.contains('Mega')])

# to exclude given string use ~
print(df.loc[~df['Name'].str.contains('Mega')])
```

```python
### Working with re library to filter data
```

```python
import re

print(df.loc[df['Type 1'].str.contains('Fire|Grass',regex = True, flags = re.I)])
# Fire|Grass - Fire or Grass
# flags = re.I -  helps ignore the character case 

# finds all the names that contains Pi
print(df.loc[df['Name'].str.contains('Pi[a-z]',flags = re.I, regex = True)])      

# finds all the names that  start with Pi
print(df.loc[df['Name'].str.contains('^Pi[a-z]',flags = re.I, regex = True)])          

# finds all the names that ends with with u
print(df.loc[df['Name'].str.contains('[a-z]u$',flags = re.I, regex = True)])          
```

```python
### Conditional Changes

df_changes = pd.read_csv('pokemon_data.csv')

# change value of smn from one to another --- from Fire to Flames
df_changes.loc[df_changes['Type 1'] == 'Fire', 'Type 1'] = 'Flamer'
print(df_changes[['Name','Type 1']])


# set all the pokemon as legendary who have type as Flamer
df_changes.loc[df_changes['Type 1'] == 'Flamer', 'Legendary'] =  True
print(df_changes.loc[df_changes['Type 1'] ==  'Flamer'])

#can also put conditions like df_changes['HP']>500 --- segregates the values where HP is greater than 500 
df_changes.loc[df_changes['HP']>50,['Generation','Legendary']] = [1,True]
print(df_changes.loc[df_changes['HP'] > 50])
```

```python
### Grouping data
# group the rows on the basis of Type and finding out values like mean, sum & count
print(df.groupby(['Type 1']).mean('HP')['HP'])

print(df.groupby(['Type 1']).sum('HP')['HP'])

df['count'] = 1
print(df.groupby(['Type 1','Type 2']).count()['count'])
```
